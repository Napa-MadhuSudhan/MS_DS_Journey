## Series 

In Pandas, a Series is a one-dimensional labeled array that can hold data of various types, similar to a Python list or array. The primary difference is that a Series has both data and associated labels, which are referred to as the index. The index allows for fast and efficient data access and manipulation based on the labels.

To create a Series, you can use the `pd.Series()` constructor, passing a list, array, or dictionary as the data. Let's go through some examples:

Example 1: Creating a Series from a list
```python
import pandas as pd

# Creating a Series from a list
data = [10, 20, 30, 40, 50]
series_1 = pd.Series(data)

print(series_1)
```

Output:
```
0    10
1    20
2    30
3    40
4    50
dtype: int64
```

Example 2: Creating a Series from a dictionary
```python
import pandas as pd

# Creating a Series from a dictionary
data = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
series_2 = pd.Series(data)

print(series_2)
```

Output:
```
a    10
b    20
c    30
d    40
dtype: int64
```

Example 3: Creating a Series with a custom index
```python
import pandas as pd

# Creating a Series with a custom index
data = [100, 200, 300, 400]
custom_index = ['apple', 'banana', 'cherry', 'date']
series_3 = pd.Series(data, index=custom_index)

print(series_3)
```

Output:
```
apple     100
banana    200
cherry    300
date      400
dtype: int64
```

You can see that in the first example, a default numerical index (0, 1, 2, ...) is assigned to the Series. In the second example, the keys of the dictionary become the index of the Series. And in the third example, we define a custom index explicitly.

Once you have created a Series, you can access its elements using the index labels, perform arithmetic operations, and apply various functions for data manipulation and analysis. For example:

```python
# Accessing elements by index label
print(series_3['banana'])  # Output: 200

# Performing arithmetic operations
result = series_1 + 5
print(result)
```

Output:
```
0    15
1    25
2    35
3    45
4    55
dtype: int64
```

These are just some basic examples of creating and using a Series in Pandas. Series provide a powerful way to handle and manipulate one-dimensional labeled data efficiently in Python.

## Data Frames

A DataFrame is a two-dimensional labeled data structure in Pandas. It can be thought of as a table, similar to an Excel spreadsheet or a SQL database table. DataFrames consist of rows and columns, where each column can have a different data type.

To use DataFrames in Pandas, you first need to import the library and create or load data into it. Here's an example of creating a DataFrame from a dictionary and then listing some common functions along with examples:

```python
import pandas as pd

# Example data for the DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy', 'David', 'Eva'],
    'Age': [25, 30, 22, 28, 24],
    'Math_Score': [85, 90, 78, 95, 88],
    'Science_Score': [92, 88, 76, 98, 84]
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

Output:
```
    Name  Age  Math_Score  Science_Score
0  Alice   25          85             92
1    Bob   30          90             88
2  Cathy   22          78             76
3  David   28          95             98
4    Eva   24          88             84
```

Now, let's list some common functions you can use with DataFrames:

1. `head()` and `tail()`:
   These functions allow you to display the first few or last few rows of the DataFrame, respectively.

   Example:
   ```python
   print(df.head(3))  # Display the first 3 rows
   print(df.tail(2))  # Display the last 2 rows
   ```

2. `info()`:
   This function provides concise information about the DataFrame, such as data types and non-null counts.

   Example:
   ```python
   print(df.info())
   ```

3. `describe()`:
   This function generates a summary of descriptive statistics for each numeric column in the DataFrame.

   Example:
   ```python
   print(df.describe())
   ```

4. `shape`:
   This attribute returns the dimensions (rows, columns) of the DataFrame as a tuple.

   Example:
   ```python
   print(df.shape)  # Output: (5, 4)
   ```

5. `loc[]` and `iloc[]`:
   These accessors are used for label-based and integer-location-based indexing, respectively.

   Example:
   ```python
   print(df.loc[2])     # Access row with label '2'
   print(df.iloc[3])    # Access row at index 3
   print(df.loc[:, 'Age':'Math_Score'])  # Access specific columns by label
   print(df.iloc[:, 1:3])  # Access specific columns by integer location
   ```

6. `drop()`:
   This function is used to remove rows or columns from the DataFrame.

   Example:
   ```python
   df.drop(index=[1, 3], inplace=True)  # Remove rows with index 1 and 3
   df.drop(columns='Age', inplace=True)  # Remove the 'Age' column
   ```

7. `groupby()`:
   This function is used for grouping and aggregating data based on one or more columns.

   Example:
   ```python
   grouped_df = df.groupby('Name')['Math_Score'].mean()
   print(grouped_df)
   ```

These are just a few of the many functions available in Pandas for DataFrames. DataFrame provides a wide range of tools for data manipulation, analysis, and visualization, making it a powerful tool for data wrangling in Python.

## Data Selection and Indexing in Data Frames

Data selection and indexing are fundamental operations in Pandas that allow you to access and manipulate data within a DataFrame. Here are various types of data selection and indexing techniques along with examples for each:



````python
# 1. Selection by Column: 
import pandas as pd

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22],
    'Salary': [50000, 60000, 45000]
}
df = pd.DataFrame(data)

# Selecting a single column
print(df['Name'])

# Selecting multiple columns
print(df[['Name', 'Salary']])
````

````python
# 2. Selection by Row:

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22],
    'Salary': [50000, 60000, 45000]
}
df = pd.DataFrame(data, index = ["A","B","C"])

# Selecting a single row using loc[]
print(df.loc["A"],'\n\n')

# Selecting multiple rows using loc[]
print(df.loc["B":"C"],'\n\n')

# Selecting a single row using iloc[]
print(df.iloc[0],'\n\n')

# Selecting multiple rows using iloc[]
print(df.iloc[1:3],'\n\n')

# df.iloc[0]
````

````python
# 3. Conditional Selection (Filtering):

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22],
    'Salary': [50000, 60000, 45000]
}
df = pd.DataFrame(data)

# Using a condition to filter rows
print(df[df['Age'] > 25])
````

````python
#4. Indexing and Slicing Columns and Rows:

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22],
    'Salary': [50000, 60000, 45000]
}
df = pd.DataFrame(data)
# Indexing and Slicing using loc[] for both rows and columns
print(df.loc[0, 'Name'])  # Access a single cell
print(df.loc[:, 'Name'])  # Access a single column
print(df.loc[1:2, ['Name', 'Age']])  # Access multiple rows and columns

# Indexing and Slicing using iloc[] for both rows and columns
print(df.iloc[0, 1])  # Access a single cell
print(df.iloc[:, 0])  # Access a single column
print(df.iloc[1:3, [0, 2]])  # Access multiple rows and columns
````

````python
# 5. Conditional Indexing and Slicing:


# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22],
    'Salary': [50000, 60000, 45000]
}
df = pd.DataFrame(data)

# Select rows based on a condition using loc[]
print(df.loc[df['Age'] > 25])

# Select rows based on a condition using iloc[]
print(df.iloc[(df['Age'] > 25).values])
````

### These are various types of data selection and indexing techniques in Pandas. By using these operations, you can efficiently access and manipulate data within a DataFrame based on your specific requirements.

## Filtering

In Pandas, there are several types of filtering techniques that allow you to extract specific rows or columns from a DataFrame based on certain conditions. Here are the common types of filtering along with examples for each:


````python
# 1. Filtering using Boolean Indexing:
#    This method involves creating a Boolean condition and using it to filter the DataFrame. The condition returns a Boolean Series, and the DataFrame is filtered based on this Series.
   

import pandas as pd

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22],
    'Salary': [50000, 60000, 45000]
}
df = pd.DataFrame(data)
# Filtering using Boolean indexing
condition = df['Age'] > 25
filtered_df = df[condition]
print(filtered_df)
````

````python
# 2. Filtering using `.query()`:
#    The `.query()` method allows you to filter a DataFrame using a query string, similar to SQL queries. This method is useful when you have complex filtering conditions.
   

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22],
    'Salary': [50000, 60000, 45000]
}
df = pd.DataFrame(data)

# Filtering using .query()
filtered_df = df.query('Age > 25')
print(filtered_df)
````

````python
# 3. Filtering using `.loc[]`:
#    The `.loc[]` method is used to filter rows and columns based on labels. It allows you to specify conditions for both rows and columns.


# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22],
    'Salary': [50000, 60000, 45000]
}
df = pd.DataFrame(data)

# Filtering using .loc[]
filtered_df = df.loc[df['Age'] > 25, ['Name', 'Salary']]
print(filtered_df)
````

````python

# 4. Filtering using `.iloc[]`:
#    The `.iloc[]` method is used to filter rows and columns based on integer positions. It allows you to specify conditions for both rows and columns.


# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22],
    'Salary': [50000, 60000, 45000]
}
df = pd.DataFrame(data)

# Filtering using .iloc[]
filtered_df = df.iloc[df['Age'].idxmax()]
print(filtered_df)
````

````python
# 5. Filtering using `.isin()`:
#    The `.isin()` method allows you to filter rows based on whether values are present in a list or another DataFrame.

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22],
    'Department': ['HR', 'Engineering', 'Finance']
}
df = pd.DataFrame(data)

# Filtering using .isin()
filtered_df = df[df['Department'].isin(['HR', 'Finance'])]
print(filtered_df)
````

These are some of the common filtering techniques in Pandas. Depending on your specific requirements, you can choose the most suitable method for filtering your DataFrame.

## Merging

 In Pandas, there are four main types of merge operations that can be performed using the `pd.merge()` function when combining DataFrames. These merge types are based on different types of set theory operations. Here are the four types of merge in DataFrame:

**1. Inner Join**:
   The inner join returns only the rows where the merge key (the common column) exists in both DataFrames. Rows with non-matching keys are excluded from the result.

Example:
```python
import pandas as pd

# Example DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Cathy']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [25, 30, 22]})

# Inner join based on 'ID'
result = pd.merge(df1, df2, on='ID', how='inner')
print(result)
```
Output:
```
   ID   Name  Age
0   2    Bob   25
1   3  Cathy   30
```

**2. Outer Join (Full Outer Join)**:
   The outer join returns all rows from both DataFrames, including rows with non-matching keys. For non-matching rows, NaN values are used to fill in the missing data.

Example:
```python
# Outer join based on 'ID'
result = pd.merge(df1, df2, on='ID', how='outer')
print(result)
```
Output:
```
   ID   Name   Age
0   1  Alice   NaN
1   2    Bob  25.0
2   3  Cathy  30.0
3   4    NaN  22.0
```

**3. Left Join**:
   The left join returns all rows from the left DataFrame (the first DataFrame specified in the merge) and the matched rows from the right DataFrame. Non-matching rows in the right DataFrame will have NaN values.

Example:
```python
# Left join based on 'ID'
result = pd.merge(df1, df2, on='ID', how='left')
print(result)
```
Output:
```
   ID   Name   Age
0   1  Alice   NaN
1   2    Bob  25.0
2   3  Cathy  30.0
```

**4. Right Join**:
   The right join returns all rows from the right DataFrame (the second DataFrame specified in the merge) and the matched rows from the left DataFrame. Non-matching rows in the left DataFrame will have NaN values.

Example:
```python
# Right join based on 'ID'
result = pd.merge(df1, df2, on='ID', how='right')
print(result)
```
Output:
```
   ID   Name  Age
0   2    Bob   25
1   3  Cathy   30
2   4    NaN   22
```

These four types of merge operations give you flexibility in combining DataFrames based on common columns, and you can choose the most suitable merge type based on your data requirements.


## JOIN 

In Pandas, the `DataFrame.join()` method is used to combine two or more DataFrames based on their indexes. It is similar to merging DataFrames with `pd.merge()`, but it operates specifically on the DataFrame indexes. The `join()` method has different types of joins that can be specified using the `how` parameter. Here are the types of `DataFrame.join()`:

1. Inner Join (`how='inner'`):
   The inner join keeps only the rows with matching index values in both DataFrames.

```python
import pandas as pd

# Example DataFrames
data1 = {
    'Name': ['Alice', 'Bob', 'Cathy'],
    'Age': [25, 30, 22]
}
df1 = pd.DataFrame(data1, index=[1, 2, 3])

data2 = {
    'Salary': [50000, 60000, 45000]
}
df2 = pd.DataFrame(data2, index=[2, 3, 4])

# Inner join based on indexes
inner_joined_df = df1.join(df2, how='inner')
print(inner_joined_df)
```

2. Left Join (`how='left'`):
   The left join keeps all the rows from the left DataFrame and the matching rows from the right DataFrame. If there is no match, the resulting DataFrame will have NaN values for the right DataFrame columns.

```python
# Left join based on indexes
left_joined_df = df1.join(df2, how='left')
print(left_joined_df)
```

3. Right Join (`how='right'`):
   The right join keeps all the rows from the right DataFrame and the matching rows from the left DataFrame. If there is no match, the resulting DataFrame will have NaN values for the left DataFrame columns.

```python
# Right join based on indexes
right_joined_df = df1.join(df2, how='right')
print(right_joined_df)
```

4. Outer Join (`how='outer'`):
   The outer join keeps all rows from both DataFrames and fills in missing values with NaN for non-matching rows.

```python
# Outer join based on indexes
outer_joined_df = df1.join(df2, how='outer')
print(outer_joined_df)
```

The `DataFrame.join()` method is a convenient way to combine DataFrames based on their indexes. It is particularly useful when you want to combine DataFrames with the same or similar indexes. Choose the appropriate join type based on your specific data and requirements.

## Difference between Merge and Join

| Aspect          | Join                                          | Merge                                       |
|-----------------|-----------------------------------------------|---------------------------------------------|
| Operation       | Operates on DataFrame indexes (row labels)   | Operates on DataFrame columns and indexes   |
| Key Parameter   | No need to specify 'on' parameter            | Requires 'on' parameter to specify key(s)   |
| Default Join    | Left join                                     | Inner join                                  |
| Handles Indexes | Yes                                           | Yes                                         |
| Handles Columns | No                                            | Yes                                         |
| Joins Types     | Left, Right, Inner, Outer                     | Left, Right, Inner, Outer                   |
| Flexibility     | Limited (indexes-only)                        | More flexible (supports various columns)    |

In summary, `join` is a specific method for combining DataFrames based on their indexes, while `merge` is a more general method that can handle merging based on both columns and indexes. The choice between `join` and `merge` depends on the specific data and the desired result.

## Grouping and Aggregation

In Pandas, there are various types of data aggregation and grouping operations that you can perform on DataFrames. Here are the common types of data aggregation and grouping along with examples for each:

1. Aggregation Functions:
   Aggregation functions are used to calculate summary statistics on numeric data in a DataFrame. Common aggregation functions include `sum()`, `mean()`, `median()`, `min()`, `max()`, `count()`, `std()`, and `var()`.

Example:

```python
import pandas as pd

# Example DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 20, 15, 25, 12]
}
df = pd.DataFrame(data)

# Aggregation: Calculate the sum of 'Value' column
sum_value = df['Value'].sum()
print("Sum:", sum_value)

# Aggregation: Calculate the mean of 'Value' column
mean_value = df['Value'].mean()
print("Mean:", mean_value)

# Aggregation: Calculate the maximum value in 'Value' column
max_value = df['Value'].max()
print("Max:", max_value)
```

2. Grouping and Aggregation:
   Grouping is the process of splitting the data into groups based on a specific column or combination of columns. Aggregation is then applied to each group independently.

Example:

```python
# Grouping by 'Category' and calculating the sum of 'Value' for each group
grouped_df = df.groupby('Category')['Value'].sum()
print(grouped_df)

# Grouping by multiple columns and calculating the mean of 'Value' for each group
grouped_df = df.groupby(['Category', 'Subcategory'])['Value'].mean()
print(grouped_df)
```

3. Pivot Tables:
   Pivot tables are used to summarize and aggregate data in a tabular format, providing a cross-tabulation of the data.

Example:

```python
# Create a pivot table: Calculate the mean of 'Value' for each combination of 'Category' and 'Subcategory'
pivot_table = df.pivot_table(index='Category', columns='Subcategory', values='Value', aggfunc='mean')
print(pivot_table)
```

4. `agg()` Function:
   The `agg()` function allows you to perform multiple aggregations on different columns simultaneously.

Example:

```python
# Aggregating using multiple functions on different columns
aggregated_df = df.agg({'Value': 'sum', 'Quantity': 'max'})
print(aggregated_df)
```

5. Transformation:
   Transformation applies a function to each group of data and returns a DataFrame with the same shape as the original.

Example:

```python
# Apply a transformation to each group: subtract the group's mean from 'Value'
df['Value_Adjusted'] = df.groupby('Category')['Value'].transform(lambda x: x - x.mean())
print(df)
```

These are some of the common types of data aggregation and grouping operations in Pandas. Depending on your specific data analysis needs, you can choose the appropriate method for summarizing and aggregating your DataFrame.

````python
# 7. Pie Chart:

import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'Category': ['A', 'B', 'C', 'D'],
        'Values': [25, 40, 60, 30]}
df = pd.DataFrame(data)

# Pie chart using Matplotlib
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%', shadow=True)
plt.title('Pie Chart of Categories')
plt.show()
````

````python
# 8. Violin Plot:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'Category': ['A'] * 10 + ['B'] * 10 + ['C'] * 10,
        'Values': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55,
                   25, 27, 29, 33, 38, 39, 43, 45, 48, 50, 55,
                   5, 15, 18, 20, 25, 28, 35, 39, 40]}
df = pd.DataFrame(data)

# Violin plot using Seaborn
sns.violinplot(x='Category', y='Values', data=df)
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Violin Plot of Categories')
plt.show()
````

````python
# 9. Heatmap:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}
df = pd.DataFrame(data)

# Heatmap using Seaborn
sns.heatmap(df, annot=True, cmap='YlGnBu')
plt.title('Heatmap of DataFrame')
plt.show()
````

````python
# 10. Pair Plot:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [5, 9, 7, 11, 14],
        'C': [3, 6, 9, 12, 15]}
df = pd.DataFrame(data)

# Pair plot using Seaborn
sns.pairplot(df)
plt.show()
````
